```python
import pandas as pd

# looks at similarity between users
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_excel(io='../Datasets/Online Retail.xlsx', sheet_name='Online Retail')
df.head()
```

```python
# Filtering out null records

df=df.loc[df['Quantity'] >0]
```

```python
# Checking for nulls in each column

df.isnull().sum()
```

```python
# check for nulls in specific column

df['CustomerID'].isna().sum()
```

```python
df.loc[df['CustomerID'].isna()].head()
```

```python
# remove missing values from the dataframe

df = df.dropna(subset=['CustomerID'])
```

## Turn the data into a customer matrix

```python
# code to create pivot table of customer data

customer_item_matrix = df.pivot_table(index='CustomerID', columns='StockCode',
                                     values='Quantity', aggfunc='sum')
customer_item_matrix
```

```python
customer_item_matrix = customer_item_matrix.applymap(lambda x: 1 if x > 0 else 0)
```

```python
# compute cosine similarities between

user_user_sim_matrix = pd.DataFrame(cosine_similarity(customer_item_matrix))
user_user_sim_matrix 
```

```python
# rename index and column names

user_user_sim_matrix.columns = customer_item_matrix.index

user_user_sim_matrix['CustomerID'] = customer_item_matrix.index
user_user_sim_matrix = user_user_sim_matrix.set_index('CustomerID')

user_user_sim_matrix
```

```python
# Ranking the most similar customers - customer 12350

user_user_sim_matrix.loc[12350.0].sort_values(ascending=False)
```

```python
# retrieve products customer 12350 HAS purchased

# add 'to_numpy' for nonzero series
items_bought_by_A = set(customer_item_matrix.loc[12350.0].iloc[
    customer_item_matrix.loc[12350.0].to_numpy().nonzero()
].index)

items_bought_by_A
```

```python
# retrieve products customer 17935 has purchased

# add 'to_numpy' for nonzero series
items_bought_by_B = set(customer_item_matrix.loc[17935.0].iloc[
    customer_item_matrix.loc[17935.0].to_numpy().nonzero()
].index)
```

```python
# The items that we can recommend to B based

items_to_recommend_to_B = items_bought_by_A - items_bought_by_B

items_to_recommend_to_B
```

```python
# look at the description of these items

df.loc[
    df['StockCode'].isin(items_to_recommend_to_B),
    ['StockCode', 'Description']
].drop_duplicates().set_index('StockCode')
```

## Item-based collaberative filtering and recommendations

```python
# Uses the similarity between items instead of customer purchases

item_item_sim_matrix = pd.DataFrame(
    cosine_similarity(customer_item_matrix.T))

item_item_sim_matrix
```

```python
# To name columns and rows

item_item_sim_matrix.columns = customer_item_matrix.T.index

item_item_sim_matrix['StockCode'] = customer_item_matrix.T.index

item_item_sim_matrix = item_item_sim_matrix.set_index('StockCode')

item_item_sim_matrix
```

```python
# similar items to a certian product (23166)

top_10_similar_items = list(item_item_sim_matrix\
                           .loc[23166]\
                           .sort_values(ascending=False)
                           .iloc[:10]\
                           .index)

top_10_similar_items
```

```python
# get the descriptions of these items
# first item is the item the customer just bought
df.loc[
    df['StockCode'].isin(top_10_similar_items),
    ['StockCode', 'Description']
].drop_duplicates().set_index('StockCode').loc[top_10_similar_items]
```
